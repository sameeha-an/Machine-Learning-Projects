In this project we will be working with a fake advertising data set, indicating whether or not a particular internet user clicked on an Advertisement. We will try to create a model that will predict whether or not they will click on an ad based off the features of that user.

This data set contains the following features:

* 'Daily Time Spent on Site': consumer time on site in minutes
* 'Age': cutomer age in years
* 'Area Income': Avg. Income of geographical area of consumer
* 'Daily Internet Usage': Avg. minutes a day consumer is on the internet
* 'Ad Topic Line': Headline of the advertisement
* 'City': City of consumer
* 'Male': Whether or not consumer was male
* 'Country': Country of consumer
* 'Timestamp': Time at which consumer clicked on Ad or closed window
* 'Clicked on Ad': 0 or 1 indicated clicking on Ad

### Split the data into features and target variables (X and y)

```python
# Choose the columns you see fit
```

```python
X = df.drop(['Clicked on Ad', 'Ad Topic Line', 'City', 'Country', 'Timestamp','Year','Date','months','Hour'], axis=1)
y = df['Clicked on Ad']
```

```python
X.head()
```

```python
y.head()
```

```python
df.head()
```

### Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_scaled = scaler.fit_transform(X)
```

```python
X = pd.DataFrame(x_scaled,columns = X.columns)
```

```python
X.head()
```

### Split the data into training and testing set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25, random_state = 50)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
y_train.shape
```

### Create a Logistic Regression model and train it

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
#train the model
model.fit(X_train,y_train)
```

### Check the accuracy of our model

```python
model.score(X_train,y_train)
```

### Make prediction using the X_test

```python
y_pred = model.predict(X_test)
```

### Check how accurate the prediction is:

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

### Check the confusion matrix

```python
print('\nConfusion Matrix:')
print(confusion_matrix(y_test, y_pred))
```

### Plot the confusion matrix on a heatmap

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False, xticklabels=['Not Clicked', 'Clicked'],yticklabels=['Not Clicked', 'Clicked'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

### Create a classification report for the model

```python
report = classification_report(y_test, y_pred)
print("Classification Report:\n", report)
```
